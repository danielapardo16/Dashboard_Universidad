```python
import pandas as pd

# Cargar dataset
df = pd.read_csv("university_student_data.csv")

# Mostrar las primeras filas
df.head()
```

### **DESCRIPCION DEL DATASET**

El archivo **university_student_data.csv** contiene informacion academica y administrativa de los estudiantes de una universidad, recopilada por año y departamento.

Estos datos permitenn analizar indicadores como la retencion, la sastifaccion estudiantil y la matricula general.

| **Columna** | **Descripcion** |
| --------------|---------------|
| ´year´ | Año academico en el que se registran los datos. |
| ´ter,´ | Semestre academico (*Spring* o *Fall*). |
| ´department´ | Departamento academico donde pertenece el estudiante. |
| ´retention_rate´ | Porcentaje de estudiantes que continuan en la institucion. |
| ´student_satisfaction´ | Nivel promedio de la sastifaccion de los estudiantes. |
|´dropout_rate´ | Pocentaje de estudiantes que abandonaron los estudios durante el periodo. |
|´enrollment_count´ | Total de estudiantes matriculados en el año con su departamento correspondiente. |

```python
# Informacion general del dataset
df.info()

# Resumen estadistico de las columnas numericas
df.describe()

# Revisar si hay valores nulos
df.isnull().sum
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.lineplot(x='Year', y='Retention Rate (%)', data=df, marker='o', color='blue')
plt.title('Tendencia de la tasa de retencion por año')
plt.xlabel("Año")
plt.ylabel("Tasa de retencion (%)")
plt.grid(True)
plt.show()
```

## **Interpretacion del grafico de retencion**
En este gráfico se observa la tendencia de la tasa de retención de los estudiantes a lo largo de los años.  
La línea muestra si el porcentaje de estudiantes que permanecen en la universidad ha aumentado o disminuido con el tiempo.  
En general, una tendencia ascendente indica una mejora en la permanencia estudiantil, mientras que una caída refleja posibles problemas de deserción o retención.


```python
plt.figure(figsize=(8, 5))
sns.lineplot(x='Year', y='Student Satisfaction (%)', data=df, palette='viridis')
plt.title('Promedio de sastifaccion estudiantil por año')
plt.xlabel("Año")
plt.ylabel("Nivel promedio de sastifaccion")
plt.show()
```

## **Interpretacion del grafico de sastifaccion estudiantil** ##
Este gráfico muestra el promedio del nivel de satisfacción de los estudiantes por año académico.  
Los valores más altos representan una percepción positiva sobre la calidad educativa y los servicios institucionales.  
Comparar los años permite identificar si la satisfacción ha mejorado o si existen periodos con menor valoración.

```python
plt.figure(figsize=(6, 5))
sns.lineplot(x='Term', y='Retention Rate (%)', data=df, palette='pastel')
plt.title("Comparacion de retencion entre terminos (Spring vs Fall)")
plt.xlabel("Termino Academico")
plt.ylabel("Tasa De Retencion (%)")
plt.show()
```

## **Interpretacion del grafico de comparacion por termino** ##
En esta gráfica se comparan las tasas de retención entre los términos académicos *Spring* y *Fall*.  
Esto permite analizar si existe alguna diferencia significativa en el comportamiento de los estudiantes según el periodo.  
Por ejemplo, si la retención es mayor en *Spring*, podría relacionarse con estrategias de reingreso o periodos de matrícula más estables.
